```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

Queue_Info = 'D:\HPC_DATA\CST-2019\Done\CST-2019_User_Feature_Collection.xlsx'
queueinfo = pd.read_excel(Queue_Info)
qi = queueinfo
qi = qi[qi['Status'] == 'Active']
qi = qi[['User', 'MEAN_Pend', 'MEAN_Memory','MEAN_CPU_Time', 'MEAN_Real_Time', 'MEAN_Submmition_Count_Recorded', 'MEAN_Submmition_Count','Days_Recorded','Jobs_Recorded']]
qi.set_index(['User'], inplace=True)

qi = qi.apply(lambda x:x.astype(float))

# qi.info()
```

```python
qi_corr = qi.corr()
```

```python
#

f, ax= plt.subplots(figsize = (8, 8))
# cmap = sns.cubehelix_palette(start = 1, rot = 3, gamma=0.8, as_cmap = True)
#or cmap='RdBu_r'
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
# Generate a mask for the upper triangle
mask = np.zeros_like(qi_corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = False

sns.heatmap(qi_corr, annot=True,annot_kws={'size':12,'weight':'bold', 'color':'black'}, mask=mask,cmap=cmap,
            square=True, robust=True, center=0, linewidths=0.05, cbar_kws={"shrink": .5})

ax.set_title('Correlation between Recent Active User\'s Job Features', fontsize=20, position=(0.5,1.05))
# ax.invert_yaxis() #Y
# ax.set_xlabel('X Label',fontsize=10)
# ax.set_ylabel('Y Label',fontsize=10)
ax.tick_params(axis='both',labelsize=12) # x y
# ax.xaxis.tick_top() # Ytop
# ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
# ax.set_yticklabels(ax.get_yticklabels(), rotation=0)

plt.show()
f.savefig('sns_style_origin.jpg', dpi=100, bbox_inches='tight')
```

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 
sns.set(style="white")
 
# Generate a large random dataset
rs = np.random.RandomState(33)
d = pd.DataFrame(data=rs.normal(size=(100, 26)),
                 columns=list(ascii_letters[26:]))
 
# Compute the correlation matrix
corr = d.corr()
 
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
 
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))
 
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
 
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```
